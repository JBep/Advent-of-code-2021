We come across a bunch of lanternfish, let's model their growth rate.

Each fish creates a new one every 7 days, thus, a fish can be modeled as a single number according to how many days it's left until a new one is created. Furthermore, a new lanternfish needs two more days before it starts it's first cycle.

```python
class Lanternfish():

    def __init__(self, amount = 1, time = None):
        if time is None:
            self.timer = 8
        else:
            self.timer = time
        self.amount = amount

    def __str__(self):
        return 'This fish has {} days left until it creates a new fish.'.format(self.timer)

#fishes = [Lanternfish(int(c)) for c in open('input_day6.txt','r').read().split(',')]

# Test input:
fishes = [Lanternfish(int(c)) for c in '3,4,3,1,2'.split(',')]
```

```python
# Task 1, how many lanternfish are there after 80 days?

contents = open('input_day6.txt','r').read().split(',')
# Test input:
#contents = '3,4,3,1,2'.split(',')

# Each lanternfish-object keeps track of the fishes with a certain time.
fishes = []
for c in contents:
    added = False
    for fish in fishes:
        if fish.timer == int(c):
            fish.amount += 1
            added = True
    
    if not added:
        fishes.append(Lanternfish(time=int(c)))


def no_fish(fishes,days):
    for day in range(0,days):
        print(day+1)
        new_fish_count = 0
    
        for fish in fishes:
            if(fish.timer == 0):
                new_fish_count += fish.amount
                fish.timer = 6
            else:
                fish.timer -= 1
    
        # Only one object is added each day, making the code scale well.
        fishes.append(Lanternfish(amount = new_fish_count))

        
    no_fishes = 0
    for fish in fishes:
        no_fishes +=fish.amount

    print(f'After {days} days there are {no_fishes} fishes.')


no_fish(fishes,80)

        
```

```python
# Task 2, do it for 256 days.
contents = open('input_day6.txt','r').read().split(',')
# Test input:
#contents = '3,4,3,1,2'.split(',')

fishes = []
for c in contents:
    added = False
    for fish in fishes:
        if fish.timer == int(c):
            fish.amount += 1
            added = True
    
    if not added:
        fishes.append(Lanternfish(time=int(c)))


no_fish(fishes,256)
```
